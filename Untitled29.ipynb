```python
# ─── Cell 2: Scrape, Analyze, and Report (with empty-data guard) ────────
import time, re, urllib.parse, requests, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from bs4 import BeautifulSoup
import markdown, pdfkit
from IPython.display import Markdown, display

# ── CONFIG ──────────────────────────────────────────────────────────
SEARCH_QUERY = "machine learning OR data scientist"
COUNTRIES    = ["United Arab Emirates","Saudi Arabia","Egypt","Qatar",
                "Morocco","Jordan","Bahrain","Kuwait","Oman","Tunisia"]
PER_COUNTRY  = 10   # posts per site per country
TOP_N        = 10
OUT_DIR      = Path("reports"); OUT_DIR.mkdir(exist_ok=True)

# ── AGENT 1: WebSearchAgent ─────────────────────────────────────────
def scrape_bayt(query):
    out = []
    for c in COUNTRIES:
        url = ("https://bayt.com/en/jobs/" +
               urllib.parse.quote_plus(query) + "/" +
               urllib.parse.quote_plus(c.lower().replace(" ", "-")) + "/")
        soup = BeautifulSoup(requests.get(url, timeout=10).text, "lxml")
        cards = soup.select("div.has-pointer-d")[:PER_COUNTRY]
        for card in cards:
            t = card.select_one("a.js-job-title")
            comp = card.select_one("a.js-company-name")
            if not (t and comp): continue
            out.append({
                "title": t.text.strip(),
                "company": comp.text.strip(),
                "description": card.get_text(" ", strip=True),
                "country": c
            })
        time.sleep(1)
    return out

def scrape_wuzzuf(query):
    out = []
    for c in COUNTRIES:
        url = ("https://wuzzuf.net/search/jobs/?a=hpb&q=" +
               urllib.parse.quote_plus(query) +
               f"&l={urllib.parse.quote_plus(c)}")
        soup = BeautifulSoup(requests.get(url, timeout=10).text, "lxml")
        cards = soup.select("div.job-card")[:PER_COUNTRY]
        for card in cards:
            t = card.select_one("h2.job-card__title")
            comp = card.select_one("a.job-card__company-name")
            if not (t and comp): continue
            out.append({
                "title": t.text.strip(),
                "company": comp.text.strip(),
                "description": card.get_text(" ", strip=True),
                "country": c
            })
        time.sleep(1)
    return out

print("  Scraping Bayt + Wuzzuf …")
raw_posts = scrape_bayt(SEARCH_QUERY) + scrape_wuzzuf(SEARCH_QUERY)
print(f"  Collected {len(raw_posts)} postings")

# ── Guard against no data ───────────────────────────────────────────
if not raw_posts:
    print("  No job postings found — check selectors or network.")
else:
    # ── AGENT 2: Data Extraction ────────────────────────────────────
    SKILL_RE = re.compile(r"(python|tensorflow|pytorch|sql|spark|hadoop|docker|aws|azure|nlp|cv)", re.I)
    rows = []
    for p in raw_posts:
        skills = set(SKILL_RE.findall(p["description"]))
        rows.append({
            "title":   p["title"].lower(),
            "company": p["company"],
            "country": p["country"],
            "skills":  ";".join(skills) if skills else ""
        })
    df = pd.DataFrame(rows)
    print(f"  Structured {len(df)} rows")

    if df.empty:
        print("  After cleaning, no data remains.")
    else:
        # ── AGENT 3: Trend Analysis ────────────────────────────────
        top_roles    = df["title"].value_counts().head(TOP_N)
        skill_freq   = df["skills"].str.split(";").explode().value_counts().head(TOP_N)
        country_dist = df["country"].value_counts().head(TOP_N)

        # ── AGENT 4: Report Writing ────────────────────────────────
        def barchart(series, title, fname):
            plt.figure(figsize=(6,3))
            series.plot.bar()
            plt.title(title)
            plt.xticks(rotation=45, ha="right")
            plt.tight_layout()
            plt.savefig(OUT_DIR/fname)
            plt.close()

        barchart(top_roles,    "Top Roles",            "roles.png")
        barchart(skill_freq,   "Top Skills",           "skills.png")
        barchart(country_dist, "Country Distribution", "countries.png")

        md = f"""# Top AI/ML Jobs in MENA

**Query:** {SEARCH_QUERY}
**Total scraped:** {len(raw_posts)}

## 1 · Top {TOP_N} Roles
![roles](roles.png)

## 2 · Top {TOP_N} Skills
![skills](skills.png)

## 3 · Country Distribution
![countries](countries.png)
"""
        report_md = OUT_DIR/"report.md"
        report_md.write_text(md)
        pdfkit.from_file(str(report_md), str(OUT_DIR/"report.pdf"))
        print(f"  Report written to {report_md}")

        display(Markdown(md))
```

```python
import urllib.parse, requests
from bs4 import BeautifulSoup

SEARCH_QUERY = "machine learning OR data scientist"
COUNTRY = "United Arab Emirates"

url = (
    "https://bayt.com/en/jobs/"
    + urllib.parse.quote_plus(SEARCH_QUERY) + "/"
    + urllib.parse.quote_plus(COUNTRY.lower().replace(" ", "-")) + "/"
)
print("URL:", url)
resp = requests.get(url, timeout=10)
print("Status code:", resp.status_code)
soup = BeautifulSoup(resp.text, "lxml")

# List the first 20 distinct class strings on the page
classes = set()
for tag in soup.find_all(True):
    cls = tag.get("class")
    if cls:
        classes.add(" ".join(cls))
print("Found classes (first 20):", list(classes)[:20], "...")

# Find any tags whose class name contains “job”
job_tags = [t for t in soup.find_all(True) if any("job" in c.lower() for c in (t.get("class") or []))]
print(f"Found {len(job_tags)} tags with “job” in class; showing first 5:\n")
for t in job_tags[:5]:
    print("-", t.name, t.get("class"), repr(t.text.strip()[:80]))
```

```python
import re, requests, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
import markdown, pdfkit
from IPython.display import Markdown, display

# ── CONFIG ──────────────────────────────────────────────────────────
API_URL = "https://remoteok.com/api"
FILTER_KEYWORDS = ["machine learning", "data scientist"]
TOP_N = 10
OUT_DIR = Path("reports"); OUT_DIR.mkdir(exist_ok=True)

# ── AGENT 1: WebSearchAgent via RemoteOK API ─────────────────────────
print("  Fetching jobs from RemoteOK API…")
resp = requests.get(API_URL, headers={"User-Agent":"curl/7.64.1"}, timeout=10)
resp.raise_for_status()
jobs = resp.json()

# The first element is metadata; real jobs start from index 1
jobs = jobs[1:]
print(f"  Retrieved {len(jobs)} total jobs")

# Filter postings by title or tags matching our keywords
filtered = []
for j in jobs:
    title = j.get("position","").lower()
    tags = [t.lower() for t in j.get("tags",[])]
    if any(kw in title for kw in FILTER_KEYWORDS) or any(kw in " ".join(tags) for kw in FILTER_KEYWORDS):
        filtered.append({
            "title":       j.get("position","").strip(),
            "company":     j.get("company","").strip(),
            "location":    j.get("location","").strip() or "Remote",
            "description": j.get("description","")
        })
print(f"  Filtered down to {len(filtered)} ML/DS jobs")

# ── AGENT 2: DataExtractionAgent ────────────────────────────────────
SKILL_RE = re.compile(r"(python|tensorflow|pytorch|sql|spark|docker|aws|azure|nlp|cv)", re.I)
rows = []
for p in filtered:
    skills = set(SKILL_RE.findall(p["description"]))
    rows.append({
        "title":    p["title"].lower(),
        "company":  p["company"],
        "location": p["location"],
        "skills":   ";".join(skills)
    })
df = pd.DataFrame(rows)
print(f"  Cleaned {len(df)} rows")

if df.empty:
    raise RuntimeError("No filtered ML/DS jobs found!")

# ── AGENT 3: TrendAnalysisAgent ────────────────────────────────────
top_roles    = df["title"].value_counts().head(TOP_N)
skill_freq   = df["skills"].str.split(";").explode().value_counts().head(TOP_N)
loc_dist     = df["location"].value_counts().head(TOP_N)

# ── AGENT 4: ReportWriterAgent ──────────────────────────────────────
def barchart(series, title, fname):
    plt.figure(figsize=(6,3))
    series.plot.bar()
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(OUT_DIR/fname)
    plt.close()

barchart(top_roles,  "Top Roles",            "roles.png")
barchart(skill_freq, "Top Skills",           "skills.png")
barchart(loc_dist,   "Location Distribution","locations.png")

md = f"""# Top Remote ML/DS Jobs via RemoteOK

**Source:** RemoteOK API
**Total fetched:** {len(df)}

## 1 · Top {TOP_N} Roles
![roles](roles.png)

## 2 · Top {TOP_N} Skills
![skills](skills.png)

## 3 · Location Distribution
![locations](locations.png)
"""
(report_md := OUT_DIR/"report.md").write_text(md)
pdfkit.from_file(str(report_md), str(OUT_DIR/"report.pdf"))
print(f"  Report saved to {report_md} and report.pdf")

display(Markdown(md))
```
